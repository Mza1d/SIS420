```python
from cmath import sin
import random
import matplotlib.pyplot as pyt
import numpy as np

for i in range(1,5000):
    x = i

n = 5000
for i in range(n):
    y = random.randint(i-1,i+1)
    print(f"y{i} : {y}")

pyt.plot(x,y)
pyt.xlabel("Ao")
pyt.ylabel("Precio")
pyt.title("REGRESIN LINEAL")
pyt.show()
```
