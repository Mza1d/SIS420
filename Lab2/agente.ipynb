```python
#Importamos librerias

import random

# Datos para el laberinto
muro = "üß±"
espacio = " "
agente = "üïµÔ∏è"
espacioRecorrido = "‚ñ¨"
dado = 'üé≤'
```

```python
# Ingresamos datos por teclado
c = int(input("Digite cantidad de COLUMNAS: "))
f = int(input("Digite cantidad de FILAS: "))
n_Pared = int(input("Digite cantidad PAREDES: "))
n_dado = int(input("Digite cantidad de DADOS: "))

#calculamos numero de espacios
n_Espacio = (c*f) - n_Pared

def generar_Mapa(cc, ff):
    laberinto1 = []
    for i in range(0, ff):
        fila_laberinto = []
        for j in range(0, cc):
            fila_laberinto.append(muro)
        laberinto1.append(fila_laberinto)
    i_actual = random.randrange(ff)
    j_actual = random.randrange(cc)
    laberinto1[i_actual][j_actual] = espacio
    n_EG = 1
    while n_EG < n_Espacio:
        direccion = random.randrange(4)
        if direccion == 0 and i_actual > 0:
            i_actual -= 1
        elif direccion == 1 and i_actual < ff - 1:
            i_actual += 1
        elif direccion == 2 and j_actual > 0:
            j_actual -= 1
        else:
            if j_actual < cc - 1:
                j_actual += 1
        if laberinto1[i_actual][j_actual] == muro:
            laberinto1[i_actual][j_actual] = espacio
            n_EG += 1
    n_p = 0
    while n_p < n_dado:
        ii = random.randrange(ff)
        jj = random.randrange(cc)
        if laberinto1[ii][jj] == espacio:
            laberinto1[ii][jj] = dado
            n_p += 1
    return laberinto1
```

```python
def imprimir(l_i):
    for line in l_i:
        print(" ".join(map(str, line)))
```

```python
def Agente(lab_ag, ff, cc, n_P):
    n = 0
    intento = 1
    i = random.randrange(f)
    j = random.randrange(c)
    while lab_ag[i][j] != espacio:
        i = random.randrange(f)
        j = random.randrange(c)
    lab_ag[i][j] = agente
    i_anterior = i
    j_anterior = j
    print(" Intento " + str(intento))
    #print("    i: "+str(i)+"\n    j: "+str(j))
    imprimir(lab_ag)
    dir = ''
    while n < n_P:
        intento += 1
        direccion = random.randrange(4)
        if direccion == 0 and i > 0:
            dir = 'Arriba'
            i -= 1
        elif direccion == 1 and i < ff - 1:
            dir = 'Abajo'
            i += 1
        elif direccion == 2 and j > 0:
            dir = 'Izquierda'
            j -= 1
        elif direccion == 3 and j < cc - 1:
            dir = 'Derecha'
            j += 1
        else:
            dir = ''
        print("\n Intento " + str(intento))
        if (
            lab_ag[i][j] != muro
            and lab_ag[i][j] != agente
        ):
            if lab_ag[i][j] == dado:
                print("  El agente encontro el dado ")
                n += 1
            lab_ag[i_anterior][j_anterior] = espacioRecorrido
            i_anterior = i
            j_anterior = j
            lab_ag[i][j] = agente
        else:
            if dir == 'Arriba':
                i += 1
            elif dir == 'Abajo':
                i -= 1
            elif dir == 'Izquierda':
                j += 1
            elif dir == 'Derecha':
                j -= 1

        imprimir(lab_ag)
        
```

```python
laberinto_m = generar_Mapa(c, f)
imprimir(laberinto_m)
Agente(laberinto_m, f, c, n_dado)
```
